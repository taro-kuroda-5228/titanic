```python
def check_title(name: str):
    count = 0
    str6 = name
    str1 = str6.replace('Mr', '')
    str2 = str1.replace('Mrs','')
    str3 = str2.replace('Miss','')
    str4 = str3.replace('Master','')
    str5 = str4.replace('Ms','')
    if str5 == str6:
        count += 1
    else:
        count += 0
    return count
```
